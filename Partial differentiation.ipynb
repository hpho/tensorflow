```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt

he_init = tf.keras.initializers.he_normal()
def act_f(features):
    return tf.math.exp(features) - 1
act_f1 = tf.math.softplus
```

```python
tf.__version__
```

```python
X = tf.placeholder(tf.float32,[None,5])
Y = tf.placeholder(tf.float32,[None,1])

W1 = tf.Variable(he_init([5,5]))
b1 = tf.Variable(he_init([5]))
L1 = act_f1(tf.matmul(X,W1)+b1)
W2 = tf.Variable(he_init([5,5]))
b2 = tf.Variable(he_init([5]))
L2 = act_f1(tf.matmul(L1,W2)+b2)
W3 = tf.Variable(he_init([5,5]))
b3 = tf.Variable(he_init([5]))
L3 = act_f1(tf.matmul(L2,W3)+b3)
W4 = tf.Variable(he_init([5,5]))
b4 = tf.Variable(he_init([5]))
L4 = act_f1(tf.matmul(L3,W4)+b4)
W5 = tf.Variable(he_init([5,5]))
b5 = tf.Variable(he_init([5]))
L5 = act_f1(tf.matmul(L4,W5)+b5)
W6 = tf.Variable(he_init([5,5]))
b6 = tf.Variable(he_init([5]))
L6 = act_f1(tf.matmul(L5,W6)+b6)



W7 = tf.Variable(he_init([5,1]))
model = tf.matmul(L6,W7)

cost = tf.math.sqrt(tf.reduce_mean(tf.pow(model-Y,2)))
optimizer = tf.train.AdamOptimizer(0.001).minimize(cost)

```

```python
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)


from sklearn.model_selection import train_test_split
Data = pd.read_csv(r'D:\Desktop\case4_one_signal.csv')
x_train,x_test = train_test_split(Data, test_size=0.33, random_state=42)
y_train = pd.DataFrame(x_train.pop('y'))
y_test = pd.DataFrame(x_test.pop('y'))
```

```python
for step in range(500000):
    sess.run(optimizer,feed_dict={X:x_train,Y:y_train})
    if (step+1)%1000 == 0:
        print((step+1),'steps : ',sess.run(cost,feed_dict={X:x_train,Y:y_train}),sess.run(cost,feed_dict={X:x_train,Y:y_train}))
        
```
