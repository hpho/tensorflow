```python
def trainlist(select, remains, num_selection):
    listcsv = pd.DataFrame()
    index = []
    for i in range(len(remains)):
        index.append(i)
    listcsv['index'] = index
    listcsv['train_arg'] = 0
    listcsv['num_selection'] = num_selection
    for j in range(len(select)):
        listcsv[select[j]] = select[j]
    listcsv['addition_parameter'] = remains
    return listcsv
```
